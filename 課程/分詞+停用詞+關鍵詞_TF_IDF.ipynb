#資料預處理
先將文本進行斷詞及去停用詞

```
#讀取預斷詞之檔案
file1=open('ctbc8.txt','r')
document1=file1.read()
#document1
```

```
#採用jieba進行斷詞，並以list存成變數word11
import jieba
word11 = jieba.lcut(document1, cut_all=False)
print("|".join(word11))
```

```
#開啟停用詞檔案，並以List格式存成變數stoplist
file3=open(u'stopwords0.txt','r',encoding='UTF-8')
stops=file3.readlines()
stopslist=[]
for i in stops:
  stopslist.append(i.strip()) #加入後去除首尾空格(strip)
stopslist.append(" ")#再加入兩個停用詞
stopslist.append("\n") #因為型態是list所以用append
print(type(stopslist))
print(stopslist)
```

```
#排除停用詞之分詞結果
resultlist=[]
for word in word11:
  if word not in stopslist:
    resultlist.append(word)
print(resultlist)
```

```
#將最後結果儲存在檔案test4中
file4=open(r'test4.txt','w', encoding='UTF-8') 
list1=" ".join(resultlist)
file4.write(list1)
file4.close()
```

#關鍵詞分析
將已乾淨的分詞資料進行關鍵詞分析

#   方法一:jieba之TF-IDF

```
import jieba.analyse
file3=open('test4.txt','r')
doc2=file3.read()
#輸出文件前10個關鍵詞，默認表20個，extract_tags：為jieba.analyse中之TF-IDF套件
keywords=jieba.analyse.extract_tags(doc2,topK=10,withWeight=True) #withWeight是否一併返回關鍵詞權重值
for key, weight in keywords:
  print('%s %f'%( key, weight))
```

```
#僅針對名詞進行關鍵詞分析
data1=jieba.analyse.extract_tags(document1,topK=5,withWeight=True,allowPOS=("n")) #allowPOS選擇詞性
for tag in data1:
  print('Word=',tag[0],'TF-IDF=',tag[1])
```

#  方法二:sklearn TF-IDF語法

```
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
```

```
tfidf2 = TfidfVectorizer()
re1=tfidf2.fit_transform([doc2]) #權重
word=tfidf2.get_feature_names() #字
print(word)
print(re1)
```

```
name=[]
weight=[]
k=0
for i in word:
    name.append(i)
    weight.append(re1[0,k])
    #print("[",i,']的權重',re1[0,k])
    k=k+1
res=pd.DataFrame([name,weight]).T
res.columns=['word','weight']
res.sort_values(by='weight',ascending=False)
```
