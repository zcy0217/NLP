```
import jieba

train = []
stopwords = open('stopwords.txt','r',encoding='utf8').read()
fp = open('主題分析文章.txt','r',encoding='utf8')
for words in fp:
  seg=jieba.cut_for_search(words)
  train.append([ word for word in seg if len(word)>1 and word not in stopwords ])
print(train)
print(len(train))
```

```
from gensim import corpora
from gensim.corpora import Dictionary #用來為每個出現在文本中的單詞分配了一個獨一無二的整數編號
from gensim.models import LdaModel

word_id = corpora.Dictionary(train) #幫單詞分配編號
#print(word_id)
#for number, word_ids in word_id.iteritems():    #顯示詞的編號
  #print(number, word_ids)
```

```
word_num = [word_id.doc2bow(text) for text in train ]    #計算詞頻向量bow,(0,3)表編號0之文字出現3次
print(word_num)
```

```
#LDA模型訓練
lda = LdaModel(corpus=word_num, id2word=word_id, num_topics=3)     #corpus：詞頻向量    
print(lda)                                 
```

```
#LDA模型訓練完成後呈現各主題之所有文詞出現之機率
topics=lda.show_topics()
for topic in topics:
  print(topic)
```

```
lda_list=lda.print_topics(2) #隨機呈現2個主題
print("2個主題的詞彙分布:\n")
for topic in lda_list:
  print(topic)
```

```
#文本2屬於各主題之機率
topic_test=lda.get_document_topics(word_num)
print(topic_test[2])
```

```
lda.save("LDA_train.model") #將主題模型儲存成LDA_train.model
load_model=LdaModel.load("LDA_train.model")
```

```
load_model.print_topic(2)   #編號2的主題
```

```
#找出某一文本的主題分布
#斷詞
import jieba

train1 = []
stopwords = open('stopwords.txt','r',encoding='utf8').read()
fp1 = open('主題分析test.txt','r',encoding='utf8')
for words in fp1:
  seg=jieba.cut_for_search(words)
  train1.append([ word for word in seg if len(word)>1 and word not in stopwords ])
print(train1)
```

```
#找詞頻
word1_id = corpora.Dictionary(train1) 
word1_num = [word1_id.doc2bow(text) for text in train1 ]
print(word1_num)
```

```
#文本之主題分佈
topic_test=load_model.get_document_topics(word1_num)
print(topic_test[0])
```
