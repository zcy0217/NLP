```
file=open('ctbc81.txt','r')
doc=file.read()
doc
```

```
from textrank4zh import TextRank4Sentence
```

```
tr4s = TextRank4Sentence()
tr4s.analyze(text=doc)

print()
print( '摘要：' )
for item in tr4s.get_key_sentences(num=4):
    print(item.index, item.weight, item.sentence)  # index是句子在文本中位置，weight是此句子的重要性
```

```
tr4s = TextRank4Sentence()
tr4s.analyze(text=doc) #lower=False(默認):是否將文本轉小寫 #source='all_filters'(默認):選擇用哪種方式，視成句子最後的詞彙，並依此來計算句子間的相似度
#source='no_filter'(所有詞彙) 'no_stop_words'(刪除結巴內建的停用詞彙) ''

print()
print( '摘要：' )
for item in tr4s.get_key_sentences(num=4):
    print(item.sentence) #item.index句子在文本中的位子 index.weight句子對文本的重要性分數
```
