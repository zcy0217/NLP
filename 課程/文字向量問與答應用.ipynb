步驟一：先將檔案中已訓練好的詞向量放入名為word_vecs這個詞典中

```
from google.colab import drive
drive.mount('/content/drive')
```

```
import numpy as np # 安裝numpy

dim = 0 #設定初始維度為0
word_vecs= {} #設定初始文字向量集合為空集合
# 由雲端硬碟上傳詞向量檔案‘cna.cbow.cwe_p.tar_g.512d.0.txt’，並將檔案設定為file1
with open('/content/drive/MyDrive/NLP/data/cna.cbow.cwe_p.tar_g.512d.0.txt') as file1:
  for line in file1:  #對於file1的每一行進行分隔詞語
    tokens = line.strip().split() #依空白作為分隔詞語

    # 第一行是兩個整數，分別代表有幾個詞向量，以及詞向量維度(512維)
    if len(tokens) == 2:
      dim = int(tokens[1]) 
      continue
    
    word = tokens[0] #取出tokens[0]，即第一行為詞語(word)
    vec = np.array([ float(t) for t in tokens[1:] ]) #取出tokens[1:]，即第二行後的資料為文字向量(word vector)
    word_vecs[word] = vec #對應各詞語的文字向量 #把檔案裡的字和文字向量取出來做成字典

# 之後可以從word_vecs這個dict中取得詞向量
```

# 一、計算文字向量

```
word_vecs["金融"]
```

```
word_vecs["銀行"]
```

```
#利用Cosine值計算文字向量間的相似程度
import numpy as np
from numpy.linalg import norm #||x||
 
# 將"金融"及“銀行”兩個文字向量分別設為A, B
A = word_vecs["金融"]
B = word_vecs["銀行"]
# 計算 A, B的相似性值(cosine similarity)
cosine = np.dot(A,B)/(norm(A)*norm(B))
print("Cosine Similarity:", cosine)
```

# 二、問與答

```
import jieba
import numpy as np
dialogue = "台南市長是哪一個政黨"
answers = [
  "今天天氣是晴天",  
  "台南市長是民進黨",
  "我喜歡吃美食", 
  "我喜歡打羽球",
  "我就讀中信金融管理學院",
  "台南最有名的夜市式花園夜市"]

word_number = 0 #設定一個句子有幾個詞的初始值為0，以作為計算後續平均值之用
avg_dialogue_vec = np.zeros((dim,)) #設定dialogue的初始向量平均數為(0,0,0,...,0) #np.zeros零向量
# 對於有在word_vecs裡面的詞我們才把它取出
# 最後詞向量加總取平均，作為句子的向量表示
for word in jieba.cut(dialogue): # 利用jieba.cut 將dialogue作分詞
  if word in word_vecs: 
    avg_dialogue_vec += word_vecs[word] #將dialogue中斷詞後的詞語向量進行加總
    word_number += 1 #將dialogue中斷詞後的詞語個數進行加總，每次加1
avg_dialogue_vec /= word_number #計算dialogue的句向量=dialogue詞語向量總和/dialogue中的詞語個數

#同樣的方式針對選項(answers)計算每個句子的句向量
word_number = 0
max_indx = -1 #設定anwser句子編號的初始值為-1
max_sim = -2 #設定最大相似性的初始值為-2
# 在六個回答中，每個答句都取詞向量平均作為向量表示
# 選出與dialogue句子向量表示cosine similarity最高的句子做為選擇的答案
for indx,ans in enumerate(answers):
  avg_answer_vec = np.zeros((dim,))
  for word in jieba.cut(ans):
    if word in word_vecs:
      avg_answer_vec += word_vecs[word]
      word_number += 1
  avg_answer_vec /= word_number
  sim = np.dot(avg_dialogue_vec, avg_answer_vec) / (np.linalg.norm(avg_dialogue_vec) * np.linalg.norm(avg_answer_vec))
  print("Ans#%d: %f" % (indx+1, sim))
  #找出答句(anwser)中與問句(dialogue)相似性最高的句子
  if sim > max_sim:
    max_indx = indx+1
    max_sim = sim

print("Answer:%d" % max_indx)
print(answers[max_indx-1])
```
