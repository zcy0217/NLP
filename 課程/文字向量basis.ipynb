```
#載入word2vec的工具包
#Word2vec是用來產生詞向量的相關模型
import pandas as pd
from gensim.models import word2vec
```

# Word2Vec的計算:資料已完成預處理(斷詞完成)

```
#sentences = [["Python", "深度學習", "機器學習"], ["NLP", "深度學習", "機器學習"],["深度學習","是","機器學習","一部分"]]
sentences = [["我","會","機器學習","也學","深度學習"],["深度學習","比","機器學習","困難"],["深度學習","是","一種","機器學習"],["我","會","機器學習","也會","深度學習"],["深度學習","是","機器學習","的","一部分"],["深度學習","是","一種","機器學習","的","建模","技術"],[ "深度學習","與","機器學習","都","可以","使用","Python"], ["NLP","深度學習","機器學習","是","屬於","AI","課程"],["機器學習","可","利用","類神經網路","建模"],["深度學習","是","利用","類神經網路","建模"]]
model1 = word2vec.Word2Vec(sentences, window=2,min_count=1, size=4,sg=0)
model1["深度學習"] #輸出深度學習的向量
#window:CBOW下決定Word2Vec一次取多少詞來預測中間詞（Skip-gram的狀況是反過來的）
#min_count：詞最少要出現幾次才會被納入詞庫
#max_vocab_size：詞庫的最大詞數
#size：詞向量的維度大小 一般向量為度都很多(上百都有可能)
#iter：訓練的回數
#sg:=0為CBOW,=1為skip-gram法
```

訓練完成之文字向量的評量:similarity相似性

```
#兩個詞之間的相似度
s0=model1.similarity("深度學習","機器學習")
print("深度學習與機器學習的相似度",s0)
```

```
#兩個詞之間的相似度
s0=model1.similarity("深度學習","NLP")
print("深度學習與NLP的相似度",s0)
```

```
jieba.load_userdict(["自然語言處理","辭彙","五花八門","機器人"])
```

```
list = jieba.lcut(doc1)   
print("/".join(list))
```

```
list1=" ".join(list)
#開啟一個空白檔案，將斷完詞的文本寫入，完成後記得要關閉檔案
file3=open(r'cutfile.txt','w',encoding='UTF-8')
file3.writelines(list1)
file3.close()
```

```
file1=open('cutfile.txt','r')
```

```
#利用LineSentence把txt文件轉爲所需要的格式:一行=一句話。單詞必須經過預處理，並由空格分隔
sentences1 = word2vec.LineSentence("cutfile.txt")
```

```
model2=word2vec.Word2Vec(sentences1,window=2,min_count=1,size=6, sg=0)
model2["AI"]
```

```
#保存模型，供日後使用
model2.save("Linword2vec.model")
```

## 載入做好的模型

```
#載入以訓練好文字向量模型，以供使用
load_model=word2vec.Word2Vec.load("Linword2vec.model")
```

```
#查看有哪些字彙已完成文字向量的訓練
#load_model.wv.vocab
```

```
load_model["AI"]
```

# 衡量文字向量訓練的狀況

```
#與某詞最相近的k個詞:topn=k
ms=load_model.wv.most_similar(['AI'],topn=5)
print(ms)
```

```
#兩個詞之間的相似度
s1=load_model.wv.similarity("AI","魔術")
print(s1)
```

```
s2=load_model.wv.similarity("AI","數據")
print(s2)
```
