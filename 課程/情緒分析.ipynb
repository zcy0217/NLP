## **安裝jieba模組**

```
!pip install jieba
import jieba
import pandas as pd
```

```
#給予負向情感以負分表示
pos1=pd.read_excel('pos1.xlsx')
neg1=pd.read_excel('neg1.xlsx')
neg1['score']=-1*neg1['score']
# neg1
```

```
#將正負情感檔案合併成情感辭典
pos_dict = dict(zip(pos1['word'],pos1['score']))
neg_dict = dict(zip(neg1['word'],neg1['score']))
sentment_dict={**pos_dict,**neg_dict}
# sentment_dict
```

```
#在情緒辭典中的文字一定要分詞出來
for w in sentment_dict.keys():
  jieba.suggest_freq(w,True)
```

```
#以定義方式，設定載入文本時，針對文本先進行斷詞後，計算文本的平均情緒分數值、平均情緒幅度
def get_sentment(sent):
    tokens = jieba.lcut(sent)
    score1 = 0
    score2 = 0
    countword = 0
    for w in tokens:
        
        if w in sentment_dict.keys():
            print(w, end=" = ")
            print(sentment_dict[w])
            score1 += sentment_dict[w]
            score2 += abs(sentment_dict[w])
            countword += 1
    print("(平均情緒分數,平均情緒幅度)=")
    if countword != 0: #當countword不等於0時，即可計算平均情緒分數,平均情緒幅度
        return score1/countword, score2/countword
    else:
        return 0
```

```
#句子之情感分析
sent =u"哥哥總是我行我素，媽媽責備他，他仍把媽媽的話當耳邊風，媽媽非常生氣"
get_sentment(sent)
```

```
sent =u"他的功課不好，但是他發憤圖強，最後得到好成績，值得當我們的模範"
get_sentment(sent)
```

```
sent =u"聽完她甜美的歌聲，覺得今天將會是美好的一天"
get_sentment(sent)
```

```
#文本之情感分析
file=open('ctbc81.txt','r')
document1=file.read()
get_sentment(document1)
```
