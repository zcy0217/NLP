#ㄧ、WordCloud直接計算文件詞頻
先將資料預處理完成:將文本進行斷詞及去停用詞，完成後之檔案名稱為test4.txt

(採用wc.generate語法)

```
file1=open('test4.txt','r')
doc1=file1.read()
doc1
```

```
#開啟圖檔
import numpy as np
from PIL import Image 

sp_mask1 = np.array(Image.open('mickey.jpeg'))
sp_mask2 = np.array(Image.open('swallow.jpg'))
```

```
#產生文字雲----font_path(字體檔) repeat=True(重複填滿)  max_words(最大填滿字) 
from wordcloud import WordCloud 
wc = WordCloud(font_path="WORDTYPE.ttf", repeat=True,background_color="white",
               max_words=2000, mask=sp_mask2,colormap="viridis")

wc.generate(doc1)
```

```
wc.to_file("S1.png")  #將此文字雲儲存成檔案

plt.figure(figsize=(8,6), dpi=120) 
plt.imshow(wc)
plt.axis("off")
plt.show()
```

# Exercise1.
# 請將homework2.txt檔進行斷詞，並設定適當停用詞，再利用WordCloud繪製文字雲

```
!pip install wordcloud
!pip install jieba
import jieba.analyse as analyse
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
from wordcloud import WordCloud 
import jieba

file1=open('homework2.txt','r')
ex1=file1.read()
jieba.add_word("黃偉哲")
jieba.add_word("藝如春風")
jieba.add_word("和氣商行")
jieba.add_word("野孩子肢體劇場")
jieba.add_word("市長")
jieba.add_word("表示")
jieba.add_word("大內區")
jieba.add_word("官田區")
jieba.add_word("趙卿惠")
jieba.add_word("陳啟榮")
jieba.add_word("不只")
list1=jieba.lcut(ex1) 
list2=" ".join(list1)

file3=open(r'test.txt','w',encoding='UTF-8')
file3.writelines(list2)
file3.close()
file4=open(r'test.txt','r').read()

sp_mask1 = np.array(Image.open('dog.jpg'))
stopwords = open('stopwords_hw2.txt','r',encoding='utf8').read()
wc = WordCloud(font_path="WORDTYPE.ttf", repeat=True,background_color="white",max_words=2000, mask=sp_mask1,colormap="viridis",stopwords=stopwords)
wc.generate(file4)
wc.to_file("ex1.png") 
plt.figure(figsize=(8,6), dpi=120) 
plt.imshow(wc)
plt.axis("off")
plt.show()
```

#二、輸入關鍵詞檔案
1.利用TFIDF關鍵詞分析後再繪製文字雲

2.利用Textrank關鍵詞分析後再繪製文字雲

(採用wc.generate_from_frequencies語法)

```
import jieba.analyse

#輸出文件前20個關鍵詞，默認表20個
keywords1=jieba.analyse.extract_tags(doc1,topK=20,withWeight=True)
dic2=dict(keywords1)
dic2
```

```
wc.generate_from_frequencies(dic2)#輸入欲繪製文字雲之文件(須為辭典格式), #wc:前述設定之文字雲格式

plt.figure(figsize=(8,6), dpi=120)   #figsize(調整圖案寬和高)  dpi(參數指定繪圖對象的分辨率,即每英吋多少像素)) 
plt.imshow(wc)  #繪製圖片
plt.axis("off")   #是否顯示坐標尺
plt.show()
```

# Exercise2.
# 請將homework2.txt檔利用TF-IDF模式進行關鍵詞分析，再利用WordCloud繪製文字雲

```
import jieba.analyse

keywords1=jieba.analyse.extract_tags(file4,topK=20,withWeight=True)
dic2=dict(keywords1)
wc.generate_from_frequencies(dic2)

plt.figure(figsize=(8,6), dpi=120) 
plt.imshow(wc) 
plt.axis("off") 
plt.show()
```

#利用jieba進行textrank關鍵詞分析後再繪製文字雲

```
keywords2=jieba.analyse.textrank(doc1,topK=20,withWeight=True)
dic3=dict(keywords2)
dic3
```

```
wc.generate_from_frequencies(dic3)#輸入欲繪製文字雲之文件(須為辭典格式)

plt.figure(figsize=(5,5), dpi=120)   #figsize(調整圖案寬和高)  dpi(參數指定繪圖對象的分辨率,即每英吋多少像素)) 
plt.imshow(wc)  #繪製圖片
plt.axis("off")   #是否顯示坐標尺
plt.show()
```
