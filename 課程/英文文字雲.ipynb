# 劃出英文文字雲

```
!pip install wordcloud               #下載文字雲軟體包。
```

```
from wordcloud import WordCloud, STOPWORDS   #下載文字雲及停用詞(a,the等)  
```

```
#開啟欲進行萃取關鍵詞之文字檔
file=open('eng1.txt','r')
document1=file.read()
document1
```

```
# 使用WordCloud中之停用詞外，另再加入一個停用詞"said")
stopwords1 = set(STOPWORDS)
stopwords1.add("said") #因為型態是set所以用append
print(type(stopwords1))
```

```
#生成文字雲
wc = WordCloud(background_color="white",repeat=True,max_words=2000,stopwords=stopwords1)
wc.generate(document1)
wc.to_file("S1.png")  #將此文字雲儲存成檔案
```

```
#將圖形畫在畫板中
import matplotlib.pyplot as plt

plt.figure(figsize=(5,5), dpi=120)   #figsize(調整圖案寬和高)  dpi(參數指定繪圖對象的分辨率,即每英吋多少像素)) 
plt.imshow(wc,interpolation='gaussian')  #繪製圖片
plt.axis("off")   #是否顯示坐標尺
plt.show()
```

# 劃出自建圖案之英文文字雲

```
# 載入桌布swallow.jpg，並設定為sp_mask1
import numpy as np
from PIL import Image 
sp_mask1 = np.array(Image.open('swallow.jpg'))
```

```
# generate word cloud, mask=sp_mask1: 以sp_mask1的圖案作為文字雲之底圖(桌布)
wc=WordCloud(background_color="white",repeat=True,max_words=2000, mask=sp_mask1)
wc.generate(document1)
wc.to_file("S2.png")
```

```
# plot the word cloud
plt.figure(figsize=(5,5), dpi=120)   #figsize(調整圖案寬和高)  dpi(參數指定繪圖對象的分辨率,即每英吋多少像素)) 
plt.imshow(wc)  #繪製圖片
plt.axis("off")   #是否顯示坐標尺
plt.show()
```

# 上課練習題1:
請利用WordCloud自動生成關鍵詞模式，繪製eng2.txt之文字雲，底圖為angel.png

```
!pip install wordcloud
from wordcloud import WordCloud, STOPWORDS
file1=open('eng2.txt','r').read()
stopwords1 = set(STOPWORDS)

import numpy as np
from PIL import Image 
sp_mask1 = np.array(Image.open('angel.png'))

wc = WordCloud(background_color="white",repeat=True,max_words=2000,stopwords=stopwords1,mask=sp_mask1)
wc.generate(file1)
wc.to_file("test1.png")

plt.figure(figsize=(5,5), dpi=120)
plt.imshow(wc) 
plt.axis("off") 
plt.show()
```

#提供文字及權重之辭典文件後，繪製文字雲
例如:採用TF-IDF之關鍵詞萃取(先將停用詞移除)後，儲存在EWC.xlsx中，再透過文字權重對應方式進行文字雲繪製

```
#讀取關鍵詞之文辭與權重檔(如EWC.xlcx)
import pandas as pd
data=pd.read_excel('EWC.xlsx')
data
```

```
#將文字及權重資料轉成辭典變數
dic2=dict(zip(data['word'],data['freq']))
dic2
```

 **方法:利用wordcloud.generate_from_frequencies(辭典檔案)**

```
#生成文字雲
wc = WordCloud(background_color="white",repeat=True,max_words=1000, mask=sp_mask1)
wc.generate_from_frequencies(dic2)

plt.figure(figsize=(8,6), dpi=120) 
plt.imshow(wc)
plt.axis("off")
plt.show()
```
