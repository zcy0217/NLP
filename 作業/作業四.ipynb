```python
import codecs
text_file=codecs.open('英文新聞.txt','rU','utf-8').read()
print(text_file)
```

```python
from wordcloud import WordCloud, STOPWORDS
import numpy as np

sp_mask1 = np.array(Image.open('1.png'))
stopwords=set(STOPWORDS)
stopwords.add('will')
stopwords.add('needs')
stopwords.add('including')
stopwords.add('According')
stopwords.add('continues')
stopwords.add('continue')
stopwords.add('said')
stopwords.add('provided')

wc=WordCloud(background_color='white',repeat=True,max_words=500,stopwords=stopwords,colormap='inferno',mask=sp_mask1)
wc.generate(text_file)

wc.to_file('result.png')

plt.figure(figsize=(8,6),dpi=120)
plt.imshow(wc,interpolation='gaussian')
plt.axis("off")
plt.show()
```

從中間最大的Inventec可以看出來這篇文章跟英業達有關係

從中可以看到很多跟科技或機器有關的單字，artificial、cooling、equipped、foundry、accelerator、GPU、Intel、machine，可推測英業達有新的科技產品或是新技術
