```python
import jieba
file1=open('hw1.txt','r') #原始檔案
file1=file1.read()
word = jieba.cut(file1,cut_all=False)
print("/".join(word))
```

```python
jieba.load_userdict('hw1-2.txt') #自建辭典檔案
seg_list = jieba.lcut(file1)
print("/".join(seg_list))
```

```python
seg_list2=" ".join(seg_list)
file3=open(r'hw1-3.txt','w',encoding='UTF-8') #斷完詞之文章檔案
file3.writelines(seg_list2)
file3.close()
```

二、請針對前述文章的第一段句子進行詞性標記。

```python
import jieba.posseg as pseg
words =pseg.cut("太陽露出小半邊臉")
for w in words:
    print(w.word,w.flag)
```
