```python
from wordcloud import WordCloud, STOPWORDS
import jieba.analyse
import pandas as pd
file3=open('eng2.txt','r')
doc2=file3.read()
keywords=jieba.analyse.extract_tags(doc2,topK=2000,withWeight=True)
key=[]
weight=[]
for key1, weight1 in keywords:
  key.append(key1)
  weight.append(weight1)
res=pd.DataFrame([key,weight]).T
res.columns=['word','weight']
data=res.sort_values(by='weight',ascending=False)
data
```

```python
dic2=dict(zip(data['word'],data['weight']))
dic2
```

```python
import numpy as np
from PIL import Image 
import matplotlib.pyplot as plt
sp_mask1 = np.array(Image.open('dog.jpg'))
wc = WordCloud(background_color="white",repeat=True,max_words=1000, mask=sp_mask1)
wc.generate_from_frequencies(dic2)

plt.figure(figsize=(8,6), dpi=120) 
plt.imshow(wc)
plt.axis("off")
plt.show()
```
