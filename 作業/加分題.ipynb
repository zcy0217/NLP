```python
file1=open('小說.txt','r') #原始檔案
file1=file1.read()
jieba.add_word("陡然間")
jieba.add_word("首領")
jieba.add_word("迫於")
jieba.add_word("師父")
jieba.add_word("補償")
jieba.add_word("窮困")
jieba.add_word("難熬")
jieba.add_word("方淩青")
jieba.add_word("殘忍")
jieba.add_word("窩囊")
jieba.add_word("傢夥")
jieba.add_word("高檔")
jieba.add_word("不對勁")
jieba.add_word("震驚")
jieba.add_word("碎屍萬段")
jieba.add_word("辱罵")
jieba.add_word("次於")
jieba.add_word("話筒")
jieba.add_word("笑聲")
jieba.add_word("稀裡糊塗")
jieba.add_word("晶鑽")
jieba.add_word("稟報")
jieba.add_word("慾念")
jieba.add_word("")
word = jieba.lcut(file1,cut_all=False)
# print("/".join(word))
```

```python
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
```

```python
x_train = [] #設x_train為空陣列
for i, text in enumerate(word): #在docs的每個句子，取出句子的編號i及句子的內容text
  word_list = text.split(' ') #設word_list為text中分出了的詞語
  a = len(word_list) #設定a為word_list句子的詞語個數
  word_list[a - 1] = word_list[a - 1].strip() #因python由0開始編號，因此最後一個詞語為a-1，而word_list[a - 1].strip()代表刪除最後一個詞語的空格
  x_train.append(word_list) #將word_list的詞語加到x_train
documents = [TaggedDocument(doc, [i]) for i, doc in enumerate(x_train)]
```

```python
model1 = Doc2Vec(documents,dm=0,vector_size=10, window=2, min_count=1)
```

```python
import numpy as np
s1 = ["台灣","是","哪個","洲"]
answers = [
  ["今天","天氣","是","晴天"],
  ["台南","市長","是","民進黨"],
  ["台灣","在","亞洲"],
  ["我","住在","台灣"]]

s1_vec=model1.infer_vector(s1)
s1_vec
```

```python
max_sim = -2
for idx,ans in enumerate(answers):
  s2_vec=model1.infer_vector(ans)
  sim = np.dot(s1_vec,s2_vec)/(np.linalg.norm(s1_vec) * np.linalg.norm(s2_vec))
  print("Ans#%d: %f" % (idx+1, sim))
  if sim > max_sim:
    max_idx = idx+1
    max_sim = sim

print("Answer:%d" % max_idx)
print(answers[max_idx-1])
```
