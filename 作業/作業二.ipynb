1.請設定適當之停用詞，將檔案homework2.txt利用TF-IDF模式(jieba和sklearn兩種語法)進行關鍵詞分析，分析出前十個關鍵詞，並列出此關鍵詞之權重。

```python
import jieba
homework2_file=open('homework2.txt','r')
homework2_file=homework2_file.read()
jieba.load_userdict(["黃偉哲","藝如春風","和氣商行","發送","到場","趙卿惠","陳啟榮","郭國文","嘉南藥理大學"])
word1=jieba.lcut(homework2_file)
print("|".join(word1))
```

```python
file4=open(r'hw2.txt','w', encoding='UTF-8') 
list1=" ".join(word1)
file4.write(list1)
file4.close()
```

```python
stop_file=open(u'stopwords0.txt','r',encoding='UTF-8')
stops=stop_file.readlines()
stopslist=[]
for i in stops:
  stopslist.append(i.strip())
stopslist.append(" ")
stopslist.append("\n")
stopslist.append("\u3000")
resultlist=[]
for word in word1:
  if word not in stopslist:
    resultlist.append(word)
print(resultlist)
```

```python
#jieba方法
import jieba.analyse
file3=open('hw2.txt','r')
doc2=file3.read()
keywords=jieba.analyse.extract_tags(doc2,topK=10,withWeight=True) 
for key, weight in keywords:
  print('%s %f'%( key, weight))
```

```python
#sklearn方法
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
tfidf2 = TfidfVectorizer()
re1=tfidf2.fit_transform([doc2])
word=tfidf2.get_feature_names()
name=[]
weight=[]
k=0
for i in word:
    name.append(i)
    weight.append(re1[0,k])
    k=k+1
res=pd.DataFrame([name,weight]).T
res.columns=['word','weight']
ans=res.sort_values(by='weight',ascending=False)
ans[:10]
```
