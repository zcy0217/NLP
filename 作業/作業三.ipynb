```python
file1=open('財經新聞.txt','r')
document1=file1.read()
```

```python
import jieba
word11 = jieba.lcut(document1, cut_all=False)
jieba.add_word("分析師")
jieba.add_word("美股")
jieba.add_word("了結")
jieba.add_word("雙十")
jieba.add_word("國泰證期")
jieba.add_word("聯準會")
jieba.add_word("位階")
jieba.add_word("個股")
jieba.add_word("乖離")
jieba.add_word("觸及")
print("|".join(word11))
```

```python
file3=open(u'stopwords0.txt','r',encoding='UTF-8')
stops=file3.readlines()
stopslist=[]
for ww in stops:
  stopslist.append(ww.strip())
stopslist.append(" ")
stopslist.append("\n")
print(stopslist)
```

```python
resultlist=[]
for word in word11:
  if word not in stopslist:
    resultlist.append(word)
print(resultlist)  
```

```python
file4=open(r'test4.txt','w', encoding='UTF-8') 
list1=" ".join(resultlist)
file4.write(list1)
file4.close()
```

TF-IDF

```python
import jieba.analyse
file3=open('test4.txt','r')
doc2=file3.read()
keywords=jieba.analyse.extract_tags(doc2,topK=10,withWeight=True)
for key, weight in keywords:
  print('%s %.3f'%( key, weight))
```

```python
import jieba.analyse
file3=open('test4.txt','r')
doc2=file3.read()
keywords=jieba.analyse.extract_tags(doc2,topK=10,withWeight=True,allowPOS=('n','v'))
for key, weight in keywords:
  print('%s %.3f'%( key, weight))
```

TextRank

```python
data2=jieba.analyse.textrank(doc2,topK=10,withWeight=True)
for key, weight in data2:
  print('%s %.3f'%( key, weight))
```

```python
data21=jieba.analyse.textrank(doc2,topK=10,withWeight=True,allowPOS=("n","v"))
for key, weight in data21:
  print('%s %.3f'%( key, weight))
```
