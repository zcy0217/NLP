請找一則近期的智慧金融相關議題之新聞，完成斷詞及停用詞處理之文件，進行以下分析：

1.直接利用WordCloud進行分析詞頻並產生文字雲。

2.利用TF-IDF模式進行關鍵詞分析後(關鍵詞設定為名詞)，再產生文字雲。

3.利用Textrank模式進行關鍵詞分析後(關鍵詞設定為名詞)，再產生文字雲。

分析呈現後請說明分析結果。

```python
#製圖模組
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
sp_mask1 = np.array(Image.open('dog.jpg'))
```

```python
file1=open('智慧金融.txt','r')
doc1=file1.read()
doc1
```

```python
import jieba
jieba.add_word("李開復")
jieba.add_word("標註")
jieba.add_word("為何")
jieba.add_word("金融業")
jieba.add_word("投資")
jieba.add_word("資料庫")
jieba.add_word("進一步")
jieba.add_word("具備")
jieba.add_word("前所未見")
jieba.add_word("併購")
jieba.add_word("幫助")
jieba.add_word("決定")
list1=jieba.lcut(doc1) 
list2=" ".join(list1)
list2
```

1.直接利用WordCloud進行分析詞頻並產生文字雲。

```python
file3=open(r'test.txt','w',encoding='UTF-8')
file3.writelines(list2)
file3.close()
file4=open(r'test.txt','r').read()
sp_mask1 = np.array(Image.open('dog.jpg'))
stopwords = open('stopwords_智慧金融.txt','r',encoding='utf8').read()

from wordcloud import WordCloud 
wc = WordCloud(font_path="WORDTYPE.ttf", repeat=True,background_color="white",max_words=2000, mask=sp_mask1,colormap="viridis",stopwords=stopwords)
wc.generate(file4)
wc.to_file("ex1.png") 
plt.figure(figsize=(8,6), dpi=120) 
plt.imshow(wc)
plt.axis("off")
plt.show()
```

2.利用TF-IDF模式進行關鍵詞分析後(關鍵詞設定為名詞)，再產生文字雲。

```python
keywords1=jieba.analyse.extract_tags(list2,topK=20,withWeight=True,allowPOS="n")
dic2=dict(keywords1)
dic2
```

```python
wc.generate_from_frequencies(dic2)
plt.figure(figsize=(8,6), dpi=120)
plt.imshow(wc)
plt.axis("off")
plt.show()
```

3.利用Textrank模式進行關鍵詞分析後(關鍵詞設定為名詞)，再產生文字雲

```python
keywords2=jieba.analyse.textrank(list2,topK=20,withWeight=True,allowPOS="n")
dic3=dict(keywords2)
dic3
```

```python
wc.generate_from_frequencies(dic3)

plt.figure(figsize=(5,5), dpi=120) 
plt.imshow(wc) 
plt.axis("off") 
plt.show()
```
