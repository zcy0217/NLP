```python
sentences1 = word2vec.LineSentence("cut.txt")
```

```python
temp1=0
temp2=0
tempi=0
tempj=0
for i in range(1,1000):
  model2=word2vec.Word2Vec(sentences1,window=2,min_count=1,size=i, sg=0)
  model2.save("CBOW.model")
  CBOW_model=word2vec.Word2Vec.load("CBOW.model")
  sim1=CBOW_model.wv.similarity("金融","銀行")
  if sim1>temp1:
    temp1=sim1
    tempi=i
for j in range(1,1000):
  model3=word2vec.Word2Vec(sentences1,window=2,min_count=1,size=j, sg=1)
  model3.save("skipgram.model")
  skipgram_model=word2vec.Word2Vec.load("skipgram.model")
  sim2=skipgram_model.wv.similarity("金融","銀行")
  if sim2>temp2:
    temp2=sim2
    tempj=j
```
