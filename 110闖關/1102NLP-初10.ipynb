```python
import codecs
text_file=codecs.open('英文新聞.txt','rU','utf-8').read()
print(text_file)
```

```python
from wordcloud import WordCloud, STOPWORDS
print(STOPWORDS)
```

```python
stopwords=set(STOPWORDS)
stopwords.add('World Bank')

wc=WordCloud(background_color='white',repeat=True,max_words=2000,stopwords=stopwords,colormap='inferno')
wc.generate(text_file)

wc.to_file('result.png')

plt.figure(figsize=(8,6),dpi=120)
plt.imshow(wc,interpolation='gaussian')
plt.axis("off")
plt.show()
```

從圖中看到growth和China呈現在圖片上是最大的，可知道是在說明關於經濟成長率的衝擊新聞

其中包含中國、烏克蘭戰爭、美國所帶來的影響
