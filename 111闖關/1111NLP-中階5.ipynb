```python
from gensim import corpora
from gensim.corpora import Dictionary #用來為每個出現在文本中的單詞分配了一個獨一無二的整數編號
from gensim.models import LdaModel

word_id = corpora.Dictionary(train) #幫單詞分配編號
print(word_id)
# for number, word_ids in word_id.iteritems():    #顯示詞的編號
#   print(number, word_ids)
word_num = [word_id.doc2bow(text) for text in train ]    #計算詞頻向量bow,(0,3)表編號0之文字出現3次
print(word_num)
```

#num_topics=3

```python
lda = LdaModel(corpus=word_num, id2word=word_id, num_topics=3)        
lda_list=lda.print_topics(3)
print("3個主題的詞彙分布:\n")
for topic in lda_list:
  print(topic)
lda.save("LDA_train.model3") #將主題模型儲存成LDA_train.model
load_model3=LdaModel.load("LDA_train.model3")
word3 = []
weight3 = []
for index, topic in load_model3.show_topics(formatted=False, num_words= 10):
    print('Topic: {} \nWords: {}'.format(index, [w[0] for w in topic]))
    word3.append([w[0] for w in topic])
    weight3.append([w[1] for w in topic])
```

#文字雲

```python
import numpy as np
from PIL import Image 

sp_mask1 = np.array(Image.open('台南.jpg'))
sp_mask2 = np.array(Image.open('dog.jpg'))
sp_mask3 = np.array(Image.open('mickey.jpeg'))
sp_mask4 = np.array(Image.open('swallow.jpg'))
```

主題一

```python
new_key = {}
for i in range(0,1):
  for j in range(10):
    if word4[i][j] in new_key:
      if new_key[word4[i][j]] > weight4[i][j]:
        pass
      elif new_key[word4[i][j]] == weight4[i][j]:
        pass
      else:
        new_key[word4[i][j]] = weight4[i][j]
    else:
      new_key[word4[i][j]] = weight4[i][j]
from wordcloud import WordCloud 
import matplotlib.pyplot as plt
wc = WordCloud(font_path='WORDTYPE.ttf', repeat=True,background_color="white",max_words=2000, mask=sp_mask1,colormap="viridis")

wc.generate_from_frequencies(new_key)#輸入欲繪製文字雲之文件(須為+)

plt.figure(figsize=(8,6), dpi=120)   #figsize(調整圖案寬和高)  dpi(參數指定繪圖對象的分辨率,即每英吋多少像素)) 
plt.imshow(wc, interpolation='gaussian')  #繪製圖片
plt.axis("off")   #是否顯示坐標尺
plt.show()
```
